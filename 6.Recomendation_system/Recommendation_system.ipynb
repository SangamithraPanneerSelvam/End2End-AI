```python
import tkinter
from tkinter import *
from tkinter import ttk

class tk():
    def __init__(self):
        self.in_season=''
        self.in_priority1=''
        self.in_priority2=''
        
    def get_inputs(self):
        in_season=''
        in_priority1=''
        in_priority2=''

        print('------------------------------------------------------------------------------------------------')
        print('Perfect Wedding')
        print('------------------------------------------------------------------------------------------------')
        print('Now Enter prefered wedding season and priorities to get possible estimations and locations:')

        OPTIONS_season = ['Summer Wedding','Spring Wedding','Fall Wedding','Winter Wedding'] #etc
        OPTIONS_priority1 = ['Attire & Accessories (Wedding Dress Suit etc.)',
         'Event Location/Venue', 'Entertainment (DJ Live Band Musician/s etc.)',
         'Bakery (Wedding Cake/dessert)' ,'Event Food Service Catering Bar Service',
         'Ceremony Officiator (Officiant Celebrant etc.)',
         'Beauty Health & Spa (Hair Makeup etc.)', 'Flowers & Decorations',
         'Event Rentals' ]
        


        master = Tk()
        master.geometry( "800x800" )

        season = StringVar(master)
        season.set(OPTIONS_season[0]) # default value

        priority1  = StringVar(master)
        priority1.set(OPTIONS_priority1[0])

    


        label_season=ttk.Label( text='Enter your favorite season :')
        label_season.pack(fill=tkinter.X, padx=5, pady=5)

        w1 = OptionMenu(master, season, *OPTIONS_season)
        w1.pack()

        label_priority1=ttk.Label( text='Enter your first priority:')
        label_priority1.pack(fill=tkinter.X, padx=5, pady=5)

        w2 = OptionMenu(master, priority1, *OPTIONS_priority1)
        w2.pack()


        def print_answers():
            self.in_season=season.get()    
            self.in_priority1=priority1.get()
         



        submit_button = tkinter.Button(master, text='Submit', command=print_answers)
        submit_button.pack()

        master.mainloop() 
        
        self.perfect_wedding()
        
    def perfect_wedding(self):
        
        cursor = conn.cursor()        
        cursor.execute("SELECT [Location],[Wedding_destination],[Describing_words],[Total_Estimate] FROM WRK_WeddingProject WHERE [Season] = '{}' AND [Priority_1] = '{}'".format(self.in_season,self.in_priority1) )

        for i in cursor:
            if i:
                print(i)

        
```
