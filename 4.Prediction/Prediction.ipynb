```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
#Loading dataset
dataset = pd.read_csv('processed_data.csv')
wedding_dataset = dataset[["Season", "Wedding_destination","Location","Priority_1","Guest_number","Total_Estimate"]]
wedding_dataset


```

```python
from re import I
from sklearn.preprocessing import LabelEncoder
le_0= LabelEncoder()
le_1= LabelEncoder()
le_2= LabelEncoder()
le_3= LabelEncoder()

wedding_dataset["Season"]=le_0.fit_transform(wedding_dataset["Season"])
wedding_dataset["Wedding_destination"]=le_1.fit_transform(wedding_dataset["Wedding_destination"])
wedding_dataset["Location"]=le_2.fit_transform(wedding_dataset["Location"])
wedding_dataset["Priority_1"]=le_3.fit_transform(wedding_dataset["Priority_1"])
wedding_dataset
#   wedding_dataset[["Season", "Wedding_destination","Location","Priority_1"]] = wedding_dataset[["Season", "Wedding_destination","Location","Priority_1"]].apply(le.fit_transform)
# wedding_dataset
```

```python
X = wedding_dataset.iloc[:, :-1].values
y = wedding_dataset.iloc[:, -1].values
```

```python
#Train, Test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# X_train =X_train.reshape(-1, 1)
# y_train =y_train.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
#display adjusted R-squared
adjusted_R2=1 - (1-regressor.score(X, y))*(len(y)-1)/(len(y)-X.shape[1]-1)
adjusted_R2
```
