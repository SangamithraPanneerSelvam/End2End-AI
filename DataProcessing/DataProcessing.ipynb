```python
import pandas as pd
dataset=pd.read_csv("wedding_info.csv")
```

```python
"""To replace the fillers and remove whitespace"""
def imputing(lis=list):
  for i in lis:
    df=dataset.replace(i,'',regex=True)

  df=df.apply(lambda x:x.str.strip() if x.dtype == "object" else x)
  return df
  

df=imputing(['\n','\t',','])
df
```

```python
"""To check for Nan"""
df.isna().sum()
```

```python
"""To replace Nan as None in string columns and replace Nan as $0 in number columns"""
df.iloc[:,0:10]=df.iloc[:,0:10].fillna("None")
df.iloc[:,10:20]=df.iloc[:,10:20].fillna('$0')
df
  
```

```python
"""Estimate column is the total amount of money for the wedding. Hence we just need to extract the amount and remove the strings. later we need to convert the column to int for calculations"""

df['Estimate']=df['Estimate'].map(lambda x:x.lstrip('$').rstrip(' is the Estimated Wedding Cost')).astype(int)
df['Estimate']
```

```python
"""For numerical columns we need to remove the $ sign and convert it to int for calculations"""
for column in df.columns[11:20]:
  df[column]=df[column].map(lambda x:x.lstrip('$'))
  df[column]=df[column].astype(int)
df  
```

```python
"""The Guest column consist of total guest, guest range and cost per person. Hence we will extract these information and convert them to integer"""

import re
df_new=pd.DataFrame()
for i in range(0,980):
    extract=re.findall(r'\d+', df['Guest'][i])
    print(extract)
    if len(extract)==4:
     
      df_new= df_new.append(pd.DataFrame([extract], 
      columns=["Guest_number","range_start","range_end","Cost_per_person"]))
      
    elif len(extract)==3:
      
      # df[["Guest_number","range","Cost_per_person"]].append(extract) 
      df_new= df_new.append(pd.DataFrame([extract], 
      columns=["Guest_number","range_end","Cost_per_person"]))
      df_new["range_start"] = '0'
    else:
      values=['0','0','0','0']
      df_new= df_new.append(pd.DataFrame([values], 
      columns=["Guest_number","range_start","range_end","Cost_per_person"]))
for column in df_new.columns:
  df_new[column]=df_new[column].astype(int) 

result=df.reset_index(drop=True).merge(df_new.reset_index(drop=True), left_index=True, right_index=True)
result
  
```

```python
""" Save to csv"""
result.to_csv('processed_data.csv')
```
